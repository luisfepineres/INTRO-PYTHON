```python
import numpy as np
import matplotlib.pyplot as plt

# Datos iniciales
v0 = 20 # velocidad inicial en m/s
theta = 30 # ángulo de lanzamiento en grados
g = 9.81 # aceleración gravitatoria en m/s^2
t = np.linspace(0, 3, 100) # tiempo de 0 a 3 segundos con 100 puntos

# Componentes de velocidad
vx = v0 * np.cos(theta * np.pi / 180)
vy = v0 * np.sin(theta * np.pi / 180)

# Posición del objeto en cada instante de tiempo
x = v0 * np.cos(theta * np.pi / 180) * t
y = v0 * np.sin(theta * np.pi / 180) * t - 0.5 * g * t**2

# Gráfico de la trayectoria
plt.plot(x, y)
plt.title('Trayectoria del objeto lanzado')
plt.xlabel('Distancia (m)')
plt.ylabel('Altura (m)')
plt.show()
```
