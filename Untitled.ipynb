```python
import math


def Bellamy(X, par):
    x = X[0]
    N_0 = par[0]
    Q_0 = par[1]
    Q_1 = par[2]
    s_0 = par[3]
    s_1 = par[4]
    w = par[5]
    alpha = par[6]
    mu = par[7]
    Q_sh = w/alpha
    if x >= Q_0:
        theta = 1.
    else:
        theta = 0.
    arg1_ped = -1*pow((x-Q_0),2)/ ( 2*pow(s_0,2) )
    arg2_ped = -1*alpha*(x-Q_0)
    Ped_fn = ( (1-w)*math.exp(arg1_ped)/(s_0*math.sqrt(2*math.pi)) + w*theta*alpha*math.exp(arg2_ped) )*math.exp(-1.*mu)
    Bel = 0.
    n = 0.
    arg_sig = 0.
    for i in range(1, 101):
        n = 1.*i
        arg_sig = -1*pow( (x- Q_0 - Q_sh - n*Q_1),2 )/( 2*n*pow(s_1,2) )        
        Bel += ( TMath.Poisson(n,mu) * math.exp(arg_sig) / (s_1*math.sqrt(2*math.pi*n)) )        
    Bel= N_0*(Bel+Ped_fn)
    if(s_1>0.95*Q_1):
        Bel=0
    return Bel
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Q0 = 30.0
Sigma0 = 5.0
mean = 0.8
Q1 = 50.0
Sigma1 = 13.0

nentries = int(1.0e6)

# Generate data
data = []
for i in range(nentries):
    npe = np.random.poisson(mean)
    X0 = np.random.normal(Q0, Sigma0)
    X = X0
    for j in range(npe):
        X1 = np.random.normal(Q1, Sigma1)
        X += X1
    data.append(X)

# Plot histogram
nbins = 500
min_value = 0
max_value = 500

fig, ax = plt.subplots()
n, bins, patches = ax.hist(data, bins=nbins, range=(min_value, max_value), alpha=0.75, color='black', edgecolor='green', linewidth=2)
ax.set_xlabel('Charge [AU]')
ax.set_ylabel('Entries')
ax.set_yscale('log')
ax.set_title('PMT Charge Distribution')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
L = 100 # Length of detector
N = 10000 # Number of cosmic rays
I0 = 1 # Intensity of cosmic rays at top of atmosphere

# Generate cosmic ray coordinates
x0 = np.zeros(N)
y0 = np.zeros(N)
z0 = np.zeros(N)

for i in range(N):
    # Randomly choose a point on the top hemisphere
    theta = np.random.uniform(0, np.pi/2)
    phi = np.random.uniform(0, 2*np.pi)
    x0[i] = L * np.sin(theta) * np.cos(phi)
    y0[i] = L * np.sin(theta) * np.sin(phi)
    z0[i] = L * np.cos(theta)

# Plot the cosmic rays
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x0, y0, z0, s=1, c='k')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim(-L/2, L/2)
ax.set_ylim(-L/2, L/2)
ax.set_zlim(0, L)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la simulación
n_events = 100000 # Número de eventos
length = 200 # Longitud de la caja (en cm)
bins = 200 # Número de bins para el histograma

# Generar eventos de muones cósmicos aleatorios
x = np.random.uniform(-length/2, length/2, n_events)
y = np.random.uniform(-length/2, length/2, n_events)
z = length/2 + np.random.exponential(scale=700, size=n_events) # Distribución exponencial para la profundidad

# Histograma 2D para la proyección XZ
plt.hist2d(x, z, bins=bins, cmap='viridis')
plt.colorbar()
plt.xlabel('Posición en X (cm)')
plt.ylabel('Profundidad (cm)')
plt.title('Proyección XZ')
plt.show()

# Histograma 2D para la proyección YZ
plt.hist2d(y, z, bins=bins, cmap='viridis')
plt.colorbar()
plt.xlabel('Posición en Y (cm)')
plt.ylabel('Profundidad (cm)')
plt.title('Proyección YZ')
plt.show()
```
